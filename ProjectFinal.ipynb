# Trying to Predict The Pollution For The Future

```python
import numpy as np

coef = np.polyfit(df1['2020'], df1['Air pollution'], 1)

# Generate predictions for entire series.

df1['predicted'] = pd.Series(np.polyval(coef, df1['2020']))

df1[['Air pollution', 'predicted']].plot()
```
